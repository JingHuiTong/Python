```python
import pandas as pd
import glob, os
import pickle
```

```python
PKLpath = '/Volumes/home/Research/STEP/01_Split'
Path = '/Volumes/home/Research/STEP/02_Station_result_csv'
phase = 'SKS'
# MAG = 6
notplotLIST = ['TRLG','BRNG','GUDG']

SAVEpath = '/Volumes/home/Research/STEP/05_Station_result_statistics'
YearRange= ''
```

```python
alldf = {'Station':[],
                      'Event':[],
                      'Ev_lat':[],
                      'Ev_lon':[],
                      'Depth':[],
                      'Mag':[],
                      'Phase':[],
                      'SCPhi':[],
                      'SCPhi_std':[],
                      'SCdt':[],
                      'SCdt_std':[],
                      'RCPhi':[],
                      'RCPhi_std':[],
                      'RCdt':[],
                      'RCdt_std':[],
                      'SNRQ':[],
                      'SNRT':[],
                      'CpH':[],
                      'Null':[],
                      'Quality':[]}
```

```python
def Calc_rho(RCdt, SCdt):
    rho = RCdt/SCdt 
    return rho
def Calc_Phi(RCPhi, SCPhi):
    Phi = max (abs(RCPhi-SCPhi), abs(SCPhi-RCPhi))
    if Phi > 90: Phi = 180 - Phi
    return Phi
```

```python
for STApath in glob.glob(f'{PKLpath}/*.pkl'):
    STApkl = pickle.load(open(STApath, "rb"))
    values = [value for keys, value in STApkl.items() ][0]
    
    NET = values['network']; STA = values['station']    
    for csvpath in glob.glob(f'{Path}/*/{NET}*{STA}*.csv'):
        print(csvpath)
        df_ = pd.read_csv(csvpath)
        df = df_[df_['Phase']==phase][df_['SNRQ']>=5][df_['Pick']==True]#[df['Mag']<7]
        for i in range(len(df)):
            event = df['Event'].values[i]
#             print(event)
            evlat = df['Ev_lat'].values[i]
            evlon = df['Ev_lon'].values[i]
            evdep = df['Depth'].values[i]
            evmag = df['Mag'].values[i]
            phase = df['Phase'].values[i]
            RCPhi = df['RCPhi'].values[i]
            RCDPhi= df['RCPhi_std'].values[i]
            RCdt  = df['RCdt'].values[i]
            RCDdt = df['RCdt_std'].values[i]
            SCPhi = df['SCPhi'].values[i]
            SCDPhi= df['SCPhi_std'].values[i]
            SCdt  = df['SCdt'].values[i]
            SCDdt = df['SCdt_std'].values[i]
            SNRQ  = df['SNRQ'].values[i]
            SNRT  = df['SNRT'].values[i]
            CpH   = df['CpH'].values[i] 
                 
            rho = Calc_rho(RCdt, SCdt)
            phi = Calc_Phi(RCPhi, SCPhi)

            if 25 < phi < 68 or CpH  > 0.76 :
                Null = True
                Quality = 'Good'
            elif 0.8 < rho < 1.1 and phi < 8:## non-null good condition 
                Null = False
                Quality = 'Good'
            elif 0.7 <= rho < 1.2 and phi <= 25: ## non-null FAIR condition 
                Null = False
                Quality = 'Fair'
            else: 
                Null = False
                Quality = 'Poor'
            NetSta = NET+'.'+STA
            alldf['Station'].append(NetSta)
            alldf['Event'].append(event)
            alldf['Ev_lat'].append(evlat)
            alldf['Ev_lon'].append(evlon)
            alldf['Depth'].append(evdep)
            alldf['Mag'].append(evmag)
            alldf['Phase'].append(phase)
            alldf['SCPhi'].append(SCPhi)
            alldf['SCPhi_std'].append(SCDPhi)
            alldf['SCdt'].append(SCdt)
            alldf['SCdt_std'].append(SCDdt)
            alldf['RCPhi'].append(RCPhi)
            alldf['RCPhi_std'].append(RCDPhi)
            alldf['RCdt'].append(RCdt)
            alldf['RCdt_std'].append(RCDdt)
            alldf['SNRQ'].append(SNRQ)
            alldf['SNRT'].append(SNRT)
            alldf['CpH'].append(CpH)
            alldf['Null'].append(Null)
            alldf['Quality'].append(Quality)
        
dff = pd.DataFrame(alldf)
dff.to_csv(f'{SAVEpath}/Event_SKSresult_2010-2020.csv',index=False)
            
```
